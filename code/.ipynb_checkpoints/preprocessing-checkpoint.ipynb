# AML Project - Preprocessing stage

**Guideliness:**
1. treatment of lost values (missing values)
2. treatment of anomalous values (outliers)
3. treatment of incoherent or incorrect values 
4. elimination of irrelevant variables
5. (possible) elimination of redundant variables
6. coding of non-continuous or non-ordered variables (nominal or binary)
7. extraction of new variables that can be useful
8. normalization of the variables (e.g. standardization)
9. transformation of the variables (e.g. correction of skewness and/or kurtosis)

```python
# This method replaces the numerical values identifying categories by 
# the tags of each category contained in 'categories.txt' and 
# 'categories-lesion.txt'
def replace_numbers_with_categories(df, standard = True, lesion = True):
    replacements = []
    # Loading replacements of standard categories
    if standard:
        with open('./data/categories.txt', mode = 'r') as file:
            for line in file: 
                var, categories = line.strip().split(':')
                if var in df:
                    replacements.append((var, categories.split()))
    # Loading replacements of derived categories (lesion)
    if lesion:
        with open('./data/categories-lesion.txt', mode = 'r') as file:
            for line in file: 
                var, categories = line.strip().split(':')
                if var in df:
                    replacements.append((var, categories.split()))
    # Replacing categories
    for var, categories in replacements:
        try:
            _ = [ float(x) for x in df[var].value_counts().index ]
        except Exception:
            print('Warning: variable "%s" not converted.' % var,
                  'Some value is non-numeric.')
            pass
        try:
            values = [ i for i in range(len(categories)) ]
            to_replace = dict(zip(values, categories))
            df[var].replace(to_replace, inplace = True)
            df[var] = df[var].astype(object)
        except:
            print('Warning: some value cannot be associated with any category.',
                  '#values(%d) != #categories(%d)' % (len(values), len(categories)))


# This method replaces the categorical values identifying categories by 
# unique integers that identify each contained in 'categories.txt' and 
# 'categories-lesion.txt'
def replace_categories_with_numbers(df, standard = True, lesion = True):
    replacements = []
    # Loading replacements of standard categories
    if standard:
        with open('./data/categories.txt', mode = 'r') as file:
            for line in file: 
                var, categories = line.strip().split(':')
                if var in df:
                    replacements.append((var, categories.split()))
    # Loading replacements of derived categories (lesion)
    if lesion:
        with open('./data/categories-lesion.txt', mode = 'r') as file:
            for line in file: 
                var, categories = line.strip().split(':')
                if var in df:
                    replacements.append((var, categories.split()))
    # Replacing categories
    for var, categories in replacements:
        to_replace = {cat:idx for idx, cat in enumerate(categories)}
        df[var].replace(to_replace, inplace = True)
        df[var] = df[var].astype(object)
        

# Returns the value counts of a Pandas Series
def counts(s, dropna = False, frame = False, maxOut = 10):
    cs = s.value_counts(dropna = False)
    if frame:
        cs = cs.to_frame()
    name = cs.name
    div = '-'*(6+len(name))
    print(name, div, sep = '\n')
    cstr = cs.to_string().split('\n')
    if len(cstr) > maxOut:
        l = int(maxOut/2)
        cstr = cstr[:l] + ['...'] + cstr[-l:]
    print('\n'.join(cstr))
    divider()
```

## Dealing with outliers

### Univariate outliers

```python
def draw_boxplot_hist(df, col, whis = 1.5):
    fig, axes= plt.subplots(1,2, gridspec_kw={'width_ratios': [1, 4]}, figsize=(9,5))
    df.boxplot(column=col, ax=axes[0], whis = whis);
    df.hist(column=col, ax=axes[1]);

def get_outliers(df, col, iqr_factor=1.5):
    # Calculating Q1, Q3 and IQR
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    # Selecting small and big outliers
    lowerBound = (q1 - iqr_factor * iqr)
    upperBound = (q3 + iqr_factor * iqr)
    small_outliers = df[col] < lowerBound
    big_outliers = df[col] > upperBound
    return small_outliers, big_outliers, lowerBound, upperBound
```

```python
iqr_factor = 3

for var in horseTrain:
    if is_numeric_dtype(horseTrain[var]):
        draw_boxplot_hist(horseTrain, var, whis = iqr_factor)
```

```python
# We store the bounds for the outlier detection for each attribute 
# in order to be able to determine outliers in the test set
outlierBounds = dict()

# Outlier detection and storage of each bound
for var in horseTrain:
    if is_numeric_dtype(horseTrain[var]):
        small, big, lower, upper = get_outliers(horseTrain, var, iqr_factor)
        outlierBounds[var] = (lower, upper)
        print('[%s outliers] small: %d, big: %d' % (var, sum(small), sum(big)))
```

```python
# Setting outliers to NA
for var in horseTrain:
    if is_numeric_dtype(horseTrain[var]):
        small, big, _, _ = get_outliers(horseTrain, var, iqr_factor)
        horseTrain.loc[small, var] = np.nan
        horseTrain.loc[big, var] = np.nan
```

## Multivariate Outliers

```python
# Method to obtain the mahalanobis distance from a set of data
# [Credits to: ]
def MahalanobisDist(data, verbose=False):
    covariance_matrix = np.cov(data, rowvar=False)
    if is_pos_def(covariance_matrix):
        inv_covariance_matrix = np.linalg.inv(covariance_matrix)
        if is_pos_def(inv_covariance_matrix):
            vars_mean = []
            for i in range(data.shape[0]):
                vars_mean.append(list(data.mean(axis=0)))
            diff = data - vars_mean
            md = []
            for i in range(len(diff)):
                md.append(np.sqrt(diff[i].dot(inv_covariance_matrix).dot(diff[i])))

            if verbose:
                print("Covariance Matrix:\n {}\n".format(covariance_matrix))
                print("Inverse of Covariance Matrix:\n {}\n".format(inv_covariance_matrix))
                print("Variables Mean Vector:\n {}\n".format(vars_mean))
                print("Variables - Variables Mean Vector:\n {}\n".format(diff))
                print("Mahalanobis Distance:\n {}\n".format(md))
            return md
        else:
            print("Error: Inverse of Covariance Matrix is not positive definite!")
    else:
        print("Error: Covariance Matrix is not positive definite!")

# Returns True if A is positive definite. Otherwise returns False
def is_pos_def(A):
    if np.allclose(A, A.T):
        try:
            np.linalg.cholesky(A)
            return True
        except np.linalg.LinAlgError:
            return False
    else:
        return False
```

```python
def get_mv_outliers(df, mdThreshold):
    # Computing Mahalanobis Distance
    numpyData = numeric(df).to_numpy()
    md = MahalanobisDist(numpyData, verbose = False)
    # Computing the individuals to remove
    dfMd = pd.concat([pd.DataFrame(md, columns = ['MDist']), horseTrain], axis = 1)
    indToRemove = dfMd[dfMd.MDist > mdThreshold] \
        .sort_values(by = ['MDist'], ascending = False)
    return indToRemove
```

```python
from scipy.stats import chi2

# Getting the MDists
numpyData = numeric(horseTrain).to_numpy()
md = MahalanobisDist(numpyData, verbose = False)
# Showing the individuals with higher MDist
dfMd = pd.concat([pd.DataFrame(md, columns=['MDist']), horseTrain], axis=1)
# Showing the MDists sorted and the cutoff we established
_ = plt.figure(figsize=(10,6), dpi=100)
_ = plt.plot(sorted(md))
_ = plt.ylabel('Mahalanobis Distance')
_ = plt.xlabel('Individuals')
## Threshold
# tsh = 6.2 
mdThreshold = chi2.ppf(0.95, horseTrain.shape[1]-1)
## First individual that surpasses the threshold
idx = next((i for i,x in enumerate(sorted(md)) if x > mdThreshold), len(md)) 
_ = plt.axvline(x = idx, color='red')
plt.show()
# Showing the individuals we will remove
indToRemove = dfMd[dfMd.MDist > mdThreshold].sort_values(by=['MDist'], ascending=False)
title('Individuals to remove: %d (tsh=%f)' % (indToRemove.shape[0], mdThreshold))

# Removing the individuals considered multivariate outliers
# horseTrain = horseTrain[np.array(md) <= tsh]
# horseTrain.reset_index(drop=True, inplace=True)

# Checking the results
# horseTrain.describe()
```

## Correction of skewness and/or kurtosis

```python
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(df, log = False):
    for var in df:
        data = np.log10(df[var]) if log else df[var]
        print('[%s]' % var)
        print('   Skewness: %.5f' % skew(data))
        print('   Kurtosis: %.5f' % kurtosis(data, fisher = False))
```

```python
from scipy import stats
from scipy.stats import shapiro

def configure_axes(ax, title):
    ax.set_title(title)
    ax.get_lines()[0].set_markerfacecolor('black')
    ax.get_lines()[0].set_markeredgecolor('black')
    ax.get_lines()[0].set_markersize(3)
    ax.get_lines()[1].set_linewidth(2)

for var in numeric(horseTrain):
    title(var)
    # Initializing plot
    fig = plt.figure(figsize = (10,10))
    
    # Probability plot
    ax1 = fig.add_subplot(221)
    prob = stats.probplot(horseTrain[var], dist = stats.norm, plot = ax1)
    configure_axes(ax1, 'Probability')
    
    # Boxcox transformation
    ax2 = fig.add_subplot(222)
    varBoxCox, la = stats.boxcox(horseTrain[var])
    prob = stats.probplot(varBoxCox, dist = stats.norm, plot = ax2)
    configure_axes(ax2, 'BoxCox')
    
    # Sqrt
    ax3 = fig.add_subplot(223)
    varSqrt = np.sqrt(horseTrain[var])
    prob = stats.probplot(varSqrt, dist = stats.norm, plot = ax3)
    configure_axes(ax3, 'SQRT')
    
    # Log
    ax4 = fig.add_subplot(224)
    varLog = np.log10(horseTrain[var])
    prob = stats.probplot(varLog, dist = stats.norm, plot = ax4)
    configure_axes(ax4, 'Log')
    # Show figures
    plt.show()
    
    # Performing Shapiro-Wilk test (for original and transformed data)
    swOrigSt, swOrigPv = shapiro(horseTrain[var])
    swBCSt, swBCPv = shapiro(varBoxCox)
    swSqrtSt, swSqrtPv = shapiro(varSqrt)
    swLogSt, swLogPv = shapiro(varLog)
    results = [('Original'.ljust(8), swOrigSt, swOrigPv),
              ('BoxCox'.ljust(8), swBCSt, swBCPv),
              ('SQRT'.ljust(8), swSqrtSt, swSqrtPv),
              ('Log'.ljust(8), swLogSt, swLogPv)]
    print('Lambda estimated for BoxCox transformation: %.5f' % la)
    print('Shapiro-Wilk test:')
    for res in results:
        print('   %s : stat = %.5f, pvalue = %.5f' % res)
```

```python
# This method plots the 'var' attribute of the 'df' DataFrame
# in original scale and log-scale (side by side)
def plot_transform(df, var, trans, bins = 20):
    fig, axes = plt.subplots(nrows = 1, ncols = 2, 
                             gridspec_kw = {'width_ratios': [2,2]}, 
                             figsize = (10,5))
    # Original scale
    df.hist(column = var, ax = axes[0], bins = bins, 
            density = True, color = 'cornflowerblue')
    sns.kdeplot(data = df, x = var, ax = axes[0], color = 'dimgray');
    axes[0].set_title(None)
    # Transform
    newvar = None
    if trans == 'log':
        newvar = np.log10(df[var])
    elif trans == 'boxcox':
        newvar, _ = stats.boxcox(df[var])
        newvar = pd.Series(newvar)
    elif trans == 'sqrt':
        newvar = np.sqrt(df[var])
    else:
        newvar = df[var]
        print('Warning: transformation not recognized.',
              'Showing the originalDF-originalDF plot.')
    newvar.hist(ax = axes[1], bins = bins, 
                density = True, color = 'lightcoral')
    plt.xlabel('%s (%s)' % (var, trans))
    sns.kdeplot(data = newvar, ax = axes[1], color = 'dimgray');
    plt.show()
```

## Discretization

```python
from sklearn.preprocessing import KBinsDiscretizer

# This methode plots the original 'col' of 'df' side-by-side with the discretized
# variable 'col' of 'dfDiscr'
def plot_discretized_var(df, dfDiscr, col):
    # Initializing plot
    fig = plt.figure(figsize = (10, 5));

    # Numerical variable
    ax1 = fig.add_subplot(121);
    df.hist(column = col, ax = ax1, bins = 15, 
                   density = False, color = 'cornflowerblue');
    plt.ylabel('count');
    plt.xlabel(col);
    plt.title(None);
    plt.grid(visible = None);

    # Discretized variable
    ax2 = fig.add_subplot(122);
    sns.countplot(data = dfDiscr, x = col, ax = ax2);
    plt.xlabel(col + ' categories');
    
    plt.show();
```

```python
# K-Means approach
# Number of categories that we will create for each numerical variable
bins = 3
# Defining the discretizer
discretizer = KBinsDiscretizer(n_bins = bins, 
                               encode = 'ordinal', 
                               strategy = 'kmeans')
# Generating model
discrModel = discretizer.fit(numeric(horseTrain))
# Discretizing train set
horseTrainDiscr = discrModel.transform(numeric(horseTrain))
horseTrainDiscr = pd.DataFrame(horseTrainDiscr, 
                               columns = numeric(horseTrain).columns,
                               dtype = object)
    
# Showing results
for col in horseTrainDiscr:
    plot_discretized_var(horseTrain, horseTrainDiscr, col)

# Setting names of categories and adding variables to the final dataset
replaceDiscr = {0.0: 'low', 1.0: 'normal', 2.0: 'high'}
for col in horseTrainDiscr:
    horseTrainDiscr[col].replace(replaceDiscr, inplace = True)
    horseTrain[col+'Discr'] = horseTrainDiscr[col]

# Checking results
horseTrain.dtypes
```

## Applying preprocessing to `test` dataset

```python
def preprocess(filename, iniColnames, iniTypes, missingColsToRemove, outlierBounds, 
               naModel, missingVars, mdThreshold, boxCoxTransform, discrModel, 
               replaceDiscr, numerizedTrainVars, stdModel, minMaxModel):
    print('01. Reading data')
    # 1. Reading data
    horse = pd.read_csv(filename, delimiter = ' ', na_values = '?', 
                        names = iniColnames, dtype = iniTypes)
    make_categories_start_at_0(horse)
    
    print('    01.1. Removing rows that have NA as the outcome')
    # 1.1. Removing rows that have NA as the outcome
    idxs = horse.loc[horse.outcome.isna()].index
    for idx in idxs:
        horse.drop(idx, axis = 0, inplace = True)
    horse.reset_index(drop = True, inplace = True)
    
    print('02. Replacing numbers with categories for the categorical attributes')
    # 2. Replacing numbers with categories for the categorical attributes
    replace_numbers_with_categories(horse, lesion = False)
    
    print('03. Dropping duplicates')
    # 3. Dropping duplicates
    duplicates = horse.loc[horse.duplicated(keep = False)].index
    horse.drop(duplicates, axis = 0, inplace = True)
    horse.reset_index(drop = True, inplace = True)
    
    print('04. Removing unnecessary variables and adding new derived ones')
    # 4. Removing unnecessary variables and adding new derived ones
    # 4.1. Drop hospital, typeLesion2, typeLesion3
    horse.drop(columns = ['hospital', 'typeLesion2', 'typeLesion3'], inplace = True)
    # 4.2. Deriving from typeLesion1
    horse['lesionSite'] = horse.typeLesion1.apply(lambda x: x[:2]).astype(int).astype(object)
    horse['lesionType'] = horse.typeLesion1.apply(lambda x: x[2:3]).astype(int).astype(object)
    horse['lesionSubtype'] = horse.typeLesion1.apply(lambda x: x[3:4]).astype(int).astype(object)
    horse['lesionCode'] = horse.typeLesion1.apply(lambda x: x[4:6]).astype(int).astype(object)
    horse.drop(columns = ['typeLesion1'], inplace = True)
    
    print('05. Replacing numbers with categories for the new derived cat. atts.')
    # 5. Replacing numbers with categories for the new derived categorical atts.
    replace_numbers_with_categories(horse, standard = False)
    
    print('06. Removing variables that had +40% of outliers in the training set')
    # 6. Removing variables that had +40% of outliers in the training set
    #    (nasogastricReflexPH, abdomAppearance, abdomProtein)
    horse.drop(missingColsToRemove, axis = 1, inplace = True)
    
    print('07. Detecting and setting to NA the outliers found using the TR set')
    # 7. Detecting and setting to NA the outliers found using the training set
    for col in outlierBounds:
        small = horse[col] < outlierBounds[col][0]
        big = horse[col] > outlierBounds[col][1]
        horse.loc[small, col] = np.nan
        horse.loc[big, col] = np.nan
    
    print('08. Removing outcome and setting categories as integer values')
    # 8. Removing outcome and setting categories as integer values
    replace_categories_with_numbers(horse)
    aux = horse.drop(['outcome'], axis = 1)
    catCols = [idx for idx, col in enumerate(aux) if not is_numeric_dtype(aux[col])]
    
    print('09. Imputing NAs with the model generated with the training set')
    # 9. Imputing NAs with the model generated with the training set
    horseImputed = naModel.transform(aux)
    horseImputed = pd.DataFrame(horseImputed, columns = aux.columns).round(1)
    horseImputed['outcome'] = horse.outcome
    
    print('10. Replacing numbers in categories as category tags (restore)')
    # 10. Replacing numbers in categories as category tags (restore)
    horseOrig = horse.copy()
    horse = horseImputed.copy()
    replace_numbers_with_categories(horse)
    
    print('11. Creation of the missing variables')
    # 11. Creation of the missing variables
    yn = {True: 'yes', False: 'no'}
    for var in missingVars:
        horse[var+'Missing'] = horseOrig[var].isna().astype(object).replace(yn)
    
    print('12. Removing multivariate outliers using the threshold set with TR data')
    # 12. Removing multivariate outliers using the threshold set with training data
    indToRemove = get_mv_outliers(horse, mdThreshold)
    if not indToRemove.empty:
        horse.drop(indToRemove.index, axis = 0, inplace = True)
    
    print('13. Apply BoxCox to %s w/ same lmbda used TR set' % \
          list(boxCoxTransform.keys()))
    # 13. Apply BoxCox to the boxCoxCols with the same lambdas used in TR set
    for col in boxCoxTransform:
        newcol = stats.boxcox(horse[col], lmbda = boxCoxTransform[col])
        horse[col] = newcol
    
    print('14. Discretization of variables')
    # 14. Discretization of variables
    horseDiscr = discrModel.transform(numeric(horse))
    horseDiscr = pd.DataFrame(horseDiscr, 
                              columns = numeric(horse).columns, 
                              dtype = object)
    for col in horseDiscr: 
        horseDiscr[col].replace(replaceDiscr, inplace = True)
        horse[col+'Discr'] = horseDiscr[col]
        
    print('15. Numerization of variables (dummy)')
    # 15. Numerization of variables (dummy)
    colsToNum = [ col for col in categoric(horse) if 'Discr' not in col and 
                 len(horse[col].value_counts().index) > 2]
    # Generating dummies
    dummies = pd.DataFrame()
    for col in colsToNum:
        aux = pd.get_dummies(horse[col], prefix = col)
        dummies = pd.concat([dummies, aux], axis = 1)
    # Checking that all dummy variables have been created
    for idx, col in enumerate(numerizedTrainVars):
        if idx < dummies.shape[1] and dummies.columns[idx] != col:
            dummies.insert(loc = idx, column = col, value = 0)
        elif idx >= dummies.shape[1]:
            dummies.insert(loc = idx, column = col, value = 0)
    # Merging DF
    horse = pd.concat([horse, dummies], axis = 1)
    
    print('16. Normalization using the scaler models generated with TR data')
    # 16. Normalization using the scaler models generated with TR data
    colsToNorm = [ col for col in numeric(horse) if '_' not in col ]
    print('    16.1. Standarization method')
    # 16.1. Standarization method
    horseStd = horse.copy()
    horseStd[colsToNorm] = stdModel.transform(horseStd[colsToNorm])
    print('    16.2. MinMax method')
    # 16.2. MinMax method
    horseMinMax = horse.copy()
    horseMinMax[colsToNorm] = minMaxModel.transform(horseMinMax[colsToNorm])
    
    return horse, horseStd, horseMinMax
```

```python
# Preprocessing test data
horseTest, horseTestStd, horseTestMinMax = preprocess('./data/horse-colic.test', 
        iniColnames, iniTypes, missingColsToRemove, outlierBounds, naModel, 
        missingVars, mdThreshold, boxCoxTransform, discrModel, replaceDiscr, 
        numerizedTrainVars, stdModel, minMaxModel)
```

## Storing results

```python
from os.path import isdir, join
from os import mkdir

# Checking output directory
path = './data/preprocessed/'
if not isdir(path):
    mkdir(path)

# Shufling and storing results in output directory (CSV format)
def store(df, path, fname):
    # Shuffling data
    df = df.sample(frac=1).reset_index(drop=True)
    # Storing data
    df.to_csv(join(path, fname), index = False)

# Train set
store(horseTrain, path, 'train.csv')
store(horseTrainStd, path, 'trainStd.csv')
store(horseTrainMinMax, path, 'trainMinMax.csv')

# Test set
store(horseTest, path, 'test.csv')
store(horseTestStd, path, 'testStd.csv')
store(horseTestMinMax, path, 'testMinMax.csv')
```
